```python
import pandas as pd
import pyodbc

from sqlalchemy import create_engine
from sqlalchemy import text
```

<span style="font-family: Calibri; font-size: 14px;"> 
Importando os arquivos CSV em pandas dataframas
</span>

```python
pnad_covid_052020 = pd.read_csv("PNAD_COVID_052020.csv", sep=",")
pnad_covid_062020 = pd.read_csv("PNAD_COVID_062020.csv", sep=",")
pnad_covid_072020 = pd.read_csv("PNAD_COVID_072020.csv", sep=",")
pnad_covid_082020 = pd.read_csv("PNAD_COVID_082020.csv", sep=",")
pnad_covid_092020 = pd.read_csv("PNAD_COVID_092020.csv", sep=",")
pnad_covid_102020 = pd.read_csv("PNAD_COVID_102020.csv", sep=",")
pnad_covid_112020 = pd.read_csv("PNAD_COVID_112020.csv", sep=",")
```

<span style="font-family: Calibri; font-size: 14px;"> 
Criando um banco de dados em SQL Server
</span>

```python
conn = pyodbc.connect(
    "DRIVER={SQL Server};SERVER=localhost\\SQLEXPRESS;DATABASE=master;Trusted_Connection=yes"
)
conn.autocommit = True 

cursor = conn.cursor()

cursor.execute("CREATE DATABASE pnad_covid_db")
print("Banco criado!")
```

<span style="font-family: Calibri; font-size: 14px;"> 
Conexão com novo banco de dados
</span>

```python
engine = create_engine(
    "mssql+pyodbc://localhost\\SQLEXPRESS/pnad_covid_db?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes", fast_executemany=True
)
```

<span style="font-family: Calibri; font-size: 14px;"> 
Carregando os dados de cada dataframe e inserindo-os em tabelas no novo banco de dados
</span>

```python
pnad_covid_052020.to_sql("pnad_covid_052020", engine, index=False, if_exists="replace", chunksize=5000) 
print("Tabela pnad_covid_052020 criada!")

pnad_covid_062020.to_sql("pnad_covid_062020", engine, index=False, if_exists="replace", chunksize=5000)
print("Tabela pnad_covid_062020 criada!")

pnad_covid_072020.to_sql("pnad_covid_072020", engine, index=False, if_exists="replace", chunksize=5000)
print("Tabela pnad_covid_072020 criada!")

pnad_covid_082020.to_sql("pnad_covid_082020", engine, index=False, if_exists="replace", chunksize=5000)
print("Tabela pnad_covid_082020 criada!")

pnad_covid_092020.to_sql("pnad_covid_092020", engine, index=False, if_exists="replace", chunksize=5000)
print("Tabela pnad_covid_092020 criada!")

pnad_covid_102020.to_sql("pnad_covid_102020", engine, index=False, if_exists="replace", chunksize=5000)
print("Tabela pnad_covid_102020 criada!")

pnad_covid_112020.to_sql("pnad_covid_112020", engine, index=False, if_exists="replace", chunksize=5000)
print("Tabela pnad_covid_112020 criada!")
```

<span style="font-family: Calibri; font-size: 14px;">
